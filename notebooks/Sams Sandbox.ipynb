```python
train = train.drop(columns=['num', 'player_id'])
test = test.drop(columns=['num', 'player_id'])
```

```python
x = pd.DataFrame(train.dtypes)
x[x[0]!="float64"]
```

```python
def convertColumnType(x, y, col, new_type):
    x[col] = x[col].astype(new_type)
    y[col] = y[col].astype(new_type)
    return x, y
```

```python
train, test = convertColumnType(train, test, 'year', 'str')

int_cols = ['GP','FTM','FTA','twoPM','twoPA','TPM','TPA']

for i in range(len(int_cols)):
    train, test = convertColumnType(train, test, int_cols[i], 'float')
```

```python
x = pd.DataFrame(train.dtypes)
x[x[0]!="float64"]
```

```python
train.describe(include='all')
```

```python
cleaned_df = train.copy()
target = cleaned_df.pop('drafted')
#X_test = test.copy()
```

```python
num_cols = list(cleaned_df.select_dtypes('number').columns)
cat_cols = list(set(cleaned_df.columns) - set(num_cols))
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False, drop='first')
```

```python
features = ohe.fit_transform(cleaned_df[cat_cols])
```

```python
features = pd.DataFrame(features, columns=ohe.get_feature_names_out())
```

```python
scaler = StandardScaler()
```

```python
features[num_cols] = scaler.fit_transform(cleaned_df[num_cols])
```

```python
from joblib import dump
dump(ohe, '../models/ohe.joblib')
dump(scaler, '../models/scaler.joblib')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_data, X_test, y_data, y_test = train_test_split(features, target, test_size=0.2, random_state=8)
X_train, X_val, y_train, y_val = train_test_split(X_data, y_data, test_size=0.2, random_state=8)
```
